## This is how we will join data from another tools such as excel and google spreadsheets

```python
#As usuall put import such as pandas, numpy, and matplotlib
import pandas as pads
import numpy as nupy
import matplotlib.pyplot as plt
%matplotlib inline


sales = pads.read_csv("data/sales_data.csv")
```

#### What's the mean of Customer_Age?

```python
sales["Customer_Age"].mean()
```

##### Go ahead and show a density (*KDE*) and a box plot with the **Customer_Age**

```python
sales['Customer_Age'].plot(kind='kde', figsize=(14,6))
plt.title("Customer Density Plot")
plt.xlabel("Age")
```

```python
sales['Customer_Age'].plot(kind='box', vert=False, figsize=(14,6))
plt.title('Customer_Age Plot (BOX)')
plt.xlabel('Customer_Age')
```

```python
sales['Unit_Cost'].plot(kind='box', vert=False, figsize=(14,6))
plt.title('Unit_Cost Plot (BOX)')
plt.xlabel('Unit_Cost')
```

```python
ax = sales['Unit_Cost'].plot(kind='kde', figsize=(14,6))
plt.title("Unit_Cost Plot")
plt.xlabel("Unit_Cost")
ax.axvline(sales["Unit_Cost"].mean(), color='Orange')
ax.axvline(sales["Unit_Cost"].median(), color='yellow')
```

```python
ax = sales['Unit_Cost'].plot(kind='hist', figsize=(14,6))
plt.title("Unit_Cost Plot")
plt.xlabel("Unit_Cost")
ax.set_xlabel("Cost per unit")
ax.set_ylabel("Number of sales")
```
